```python
import pandas as pd

usZipCodeData = pd.read_csv(
    'http://people.bu.edu/kalathur/datasets/uszips.csv', 
    converters={'zip': lambda x: str(x)}) 
```

Q1. Show the count, mean, standard deviation, min, max, and quartiles for the population attribute 

```python
q1 = usZipCodeData['population'].describe()
q1
```

Q2. How many city zip codes have fewer than 1000 people? 

```python
usZipCodeData['zip'][usZipCodeData['population'] < 1000].count()
```

Q3. Re-assign usZipCodeData variable to have only the rows where population is 1000 or above. Change the index values of this new dataframe so that they are contiguous. 

```python
import numpy as np

usZipCodeData = usZipCodeData[usZipCodeData['population'] >= 1000].reset_index()
usZipCodeData.head(20)
```

Q11. For Massachusetts, show a scatter plot of latitute versus longitude using color as the log of the population and size as 1/25 of the density. 

```python
import matplotlib.pyplot as plt


plt.scatter(ma_StateData['lng'], 
            ma_StateData['lat'],
            label=None,
            c=np.log10(ma_StateData['population']),
            cmap='viridis',
            s=ma_StateData['density']/25,
            linewidth=0,
            alpha=0.5)

plt.axis(aspect='equal')
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.colorbar(label='log$_{10}$(population)')
plt.clim(4, 7)

plt.title('Massachusetts Cities: Area and Population');
```

Q12. For the top 75 populous zip codes in the usZipCodeData, show a pie chart with the distribution of the states and the frequencies of the zip codes in those states. (Hint: Use value_counts. Do not use any aggregate functions not yet covered). 

```python
population = usZipCodeData['population'].sort_values(ascending=False)

if False == usZipCodeData['zip'].iloc[population[0:75].index].is_unique:
    raise Exception('Error in input data:  zip codes must be unique')
else:
    states = usZipCodeData['state_id'].iloc[population[0:75].index]
    frequencies = states.value_counts()

    print()
    print("frequencies:\n{}".format(frequencies))

    plt.pie([x/frequencies.sum() for x in frequencies.values],
            labels=frequencies.index,
            autopct='%.2f%%')
    plt.axis('equal');
```
