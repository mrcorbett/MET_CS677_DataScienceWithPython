### Seaborn - Visualizing statistical relationships
- relplot() - scatter and line plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="ticks")
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
# default kind="scatter"

# https://seaborn.pydata.org/tutorial/relational.html
    
# sns.relplot() provides access to several different axes-level functions that show the relationship 
# between two variables with semantic mappings of subsets. 

sns.relplot(x="total_bill", y="tip", data=tips);   # relationship plot
```

```python
# depicts the joint distribution of two variables using a cloud of points, 
# where each point represents an observation in the dataset.

sns.scatterplot(x="total_bill", y="tip", data=tips);
```

```python
# Add a third dimension - color by a third variable (Group by)
# hue semantic - color of a point now has a meaning

sns.relplot(x="total_bill", y="tip", hue="day", data=tips);
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips);
```

```python
sns.relplot(x="total_bill", y="tip", hue="sex", data=tips);
```

```python
# hue semantic categorical in above examples (qualitative palette)
# If hue semantic is numeric, sequential palette is used for color
```

```python
current_palette = sns.color_palette()
sns.palplot(current_palette)
```

```python
sns.relplot(x="total_bill", y="tip", hue="size", data=tips);
```

```python
sequential_colors = sns.color_palette("RdPu", 8)  # red/purple palette
sns.palplot(sequential_colors)
```

```python
# semantic variable for the size of each point
# literal value not used for the area
# range of values in data units normalized into a range in area units

sns.relplot(x="total_bill", y="tip", 
            size="size", hue="size", data=tips);
```

```python
# customize area sizes range

sns.relplot(x="total_bill", y="tip", 
            size="size", sizes=(50, 300), 
            hue="size", data=tips);
```

```python
# style - vary the marker (same grouping variable)

sns.relplot(x="total_bill", y="tip", hue="time", 
            style="time", data=tips);
```

```python
# style - vary the marker (different grouping variables)

sns.relplot(x="total_bill", y="tip", hue="day", 
            style="time", data=tips);
```

```python
sns.relplot(x="total_bill", y="tip", hue="day", 
            style="time", 
            markers = {"Lunch": "s", "Dinner": "X"},  # s=square
            data=tips);
```

### Showing multiple relationships
 - when a relationship between two variables depends on more than one variable?

```python
sns.relplot(x="total_bill", y="tip", hue="day", 
            col="time", data=tips);
```

```python
sns.relplot(x="total_bill", y="tip", hue="day", 
            col="time", row="sex", data=tips);
```

```python
sns.relplot(x="total_bill", y="tip", hue="time", 
            col="day",  col_wrap=2, data=tips);
```

```python
g = sns.relplot(x="total_bill", y="tip", hue="time", size="size",
                palette=["b", "r"], sizes=(10, 100),
                col="time", data=tips);
```

### Line Plots

```python
sns.set(style="whitegrid")
np.random.seed(54321)
```

```python
df = pd.DataFrame(dict(time=np.arange(365),
                       value=np.random.randn(365)))
df.head()
```

```python
sns.relplot(x="time", y="value", kind="line", data=df);
```

```python
sns.lineplot(x="time", y="value", data=df);
```

```python
df["price"] = 250 + df["value"].cumsum()
df.head()
```

```python
sns.lineplot(x="time", y="price", data=df);
```

```python
df["mean"] = df['price'].rolling(7).mean()
df["sd"]  = df["price"].rolling(7).std()
df["upper"] = df["mean"] + (2 * df["sd"])
df["lower"] = df["mean"] - (2 * df["sd"])
df.head(12)
```

```python
ax = sns.lineplot(x="time", y="upper", data=df, label="Upper")
ax = sns.lineplot(x="time", y="mean", data=df, label = "Average")
ax = sns.lineplot(x="time", y="lower", data=df, label = "Lower")
ax.fill_between(df["time"], df['upper'], df['lower'], 
                color='cyan', alpha=0.3);
```
