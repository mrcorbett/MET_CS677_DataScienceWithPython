### Seaborn - Visualizing Data Distributions

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="ticks")
```

#### Univariate Distributions
 - distplot() - draw a histogram and file a KDE (kernel density estimate)

```python
np.random.seed(123)
x = np.random.normal(size=500)
sns.distplot(x);
```

```python
sns.distplot(x, kde=False, rug=True);
```

```python
sns.distplot(x, bins=20, kde=False, rug=True);
```

```python
sns.distplot(x, bins=20, hist=False, rug=True);
```

```python
# kdeplot provides more options

sns.kdeplot(x, shade=True);
```

```python
sns.kdeplot(x, shade=True)
sns.rugplot(x);
```

### Bivariate distributions
 - jointplot()

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips);
```

```python
# For large datasets, hexbin plot shows counts of observations

sns.jointplot(x="total_bill", y="tip", 
              kind="hex", data=tips);
```

```python
# Use kde to show contour plot

sns.jointplot(x="total_bill", y="tip", 
              kind="kde", data=tips);
```

### JointGrid

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
```

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
g = g.plot(sns.regplot, sns.distplot)
```

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
g = g.plot_joint(plt.scatter)
g.plot_marginals(sns.distplot, kde=False);
```

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
g = g.plot_joint(plt.scatter, color="m", edgecolor="white")
g.ax_marg_x.hist(tips["total_bill"], color="b", alpha=.2,
                      bins=np.arange(0, 60, 5))
g.ax_marg_y.hist(tips["tip"], color="r", alpha=.2,
                      orientation="horizontal",
                      bins=np.arange(0, 12, 1));
```
