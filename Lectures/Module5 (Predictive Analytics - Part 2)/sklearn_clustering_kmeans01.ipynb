### k-Means Clustering

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
X = np.array([[1, 2], [1, 4], [1, 0],
              [2, 1], [3, 1],
              [4, 2], [4, 4], [4, 0]])
print(X)
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)
```

```python
print(kmeans.labels_)
```

```python
centers = kmeans.cluster_centers_
print(centers)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, s=32, cmap="viridis")
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=128, alpha=0.4);
```

```python
kmeans.predict([[0, 1], [2, 2], [3, 2], [3, 3], [4, 3]])
```

```python
X
```

```python
kmeans.fit_predict(X)
```

```python
kmeans.fit_transform(X)  # distances to each of the centers  3 clusters
```

```python
from sklearn.datasets.samples_generator import make_blobs
```

```python
X,y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=0)
```

```python
print(X[:8])
```

```python
print(y_true[:8])
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_true, s = 32, cmap='viridis');
```

```python
kmeans = KMeans(n_clusters=4)

kmeans.fit(X)

y_kmeans = kmeans.predict(X)
```

```python
kmeans.labels_
```

```python
print(y_kmeans[:8])
```

```python
centers = kmeans.cluster_centers_
print(centers)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=32, cmap="viridis")
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=128, alpha=0.8);
```
