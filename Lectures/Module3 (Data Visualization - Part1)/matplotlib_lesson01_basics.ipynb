#### MATLAB-style interface

```python
x = np.linspace(0, 10, 100)

plt.figure()  # create a plot figure

# create the first of four panels and set current axis
plt.subplot(2, 2, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x))

# create the third panel and set current axis
plt.subplot(2, 2, 3)
plt.plot(x, np.exp(x))

# create the fouth panel and set current axis
plt.subplot(2, 2, 4)
plt.plot(x, 2 ** x);
```

### Object-oriented interface

```python
x = np.linspace(0, 10, 100)

# First create a grid of plots
# ax will be an array of 2 x 2 Axes objects

fig, ax = plt.subplots(2,2)

# Call plot() method on the appropriate object
ax[0,0].plot(x, np.sin(x))
ax[0,1].plot(x, np.cos(x))
ax[1,0].plot(x, np.exp(x))
ax[1,1].plot(x, 2 ** x);
```

#### Line Colors and Styles

```python
x = np.linspace(0, 10, 1000)

plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.0,0.0)); # RGB tuple, values 0 to 1
```

```python
x = np.linspace(0, 10, 1000)

plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

### Case Study - Population vs. Area

```python
cities = pd.read_csv('http://people.bu.edu/kalathur/datasets/california_cities.csv')
cities.head()
```

```python
lat, lon = cities['latd'], cities['longd']

population, area = cities['population_total'], cities['area_total_km2']

population.min(), population.max(), area.min(), area.max()
```

```python
# Scatter the points, using size and color but no label  # typically used to show area
plt.scatter(lon, lat, label=None,
            c=np.log10(population), cmap='viridis',
            s=area, linewidth=0, alpha=0.5)
plt.axis(aspect='equal')
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.colorbar(label='log$_{10}$(population)')
plt.clim(4, 7)

plt.title('California Cities: Area and Population');
```

### Case Study - Masked Scatter Plot

```python
np.random.seed(321)  # analyze this
N = 100 
r0 = 1.0 # boundary radius

x = np.random.randn(N)
y = np.random.randn(N)
r = np.sqrt(x * x + y * y)

# for size parameter
area = np.pi * (10 * np.random.randn(N))**2  # pi-r-squared
area = np.sqrt(area)

# mask into two regions
masked_area1 = np.ma.masked_where(r < r0, area)
masked_area2 = np.ma.masked_where(r >= r0, area)

plt.axis([-3, 3, -3, 3]);

plt.scatter(x, y, s=masked_area1, marker='*', color='g')
plt.scatter(x, y, s=masked_area2, marker='o', color='r')

# Show the boundary between the regions:
theta = np.arange(0, 2* np.pi , .01)
plt.plot(r0 * np.cos(theta), r0 * np.sin(theta), 'k')

plt.show()
```

### Stacked Area Plot
https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.stackplot.html

```python
x = [1, 2, 3, 4, 5]

y1 = [1, 1, 2, 3, 5]
y2 = [0, 2, 4, 6, 8]
y3 = [1, 3, 5, 7, 9]
```

```python
labels = ["Fibonacci ", "Evens", "Odds"]

fig, ax = plt.subplots()
ax.stackplot(x, y1, y2, y3, labels=labels)
ax.legend(loc='upper left')
plt.show()
```

```python
y = np.vstack([y1, y2, y3])
y
```

```python
fig, ax = plt.subplots()
ax.stackplot(x, y, labels=labels)
ax.legend(loc='upper left')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.stackplot(x, y, labels=labels, baseline="wiggle")
ax.legend(loc='upper left')
plt.show()
```
