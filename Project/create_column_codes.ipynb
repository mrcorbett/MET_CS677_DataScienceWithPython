```python
# Create a Pandas Excel writer object using XlsxWriter as the engine. 
with xlsxwriter.Workbook(column_codes_fname) as workbook:
    num_format = workbook.add_format()
    num_format.set_num_format('0')
    
    for index, key in enumerate(Neiss.DICT_CATEGORY_TRANSLATOR):
        worksheet = workbook.add_worksheet(key)
        
        # Write a dataframe to the worksheet. 
        d = Neiss.DICT_CATEGORY_TRANSLATOR[key] 
        for index, key in enumerate(d.keys()):
            worksheet.write(index, 0, key, num_format)
            worksheet.write(index, 1, d[key])
```
