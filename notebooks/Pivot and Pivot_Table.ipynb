# Pivot

Pivot allows you to transform or reshape data

```python
import pandas as pd

df = pd.DataFrame({
    'date' : [
        '5/1/2017', '5/2/2017', '5/3/2017', 
        '5/1/2017', '5/2/2017', '5/3/2017',
        '5/1/2017', '5/2/2017', '5/3/2017'],

    'city' : [
        'new york', 'new york', 'new york',
        'mumbai', 'mumbai', 'mumbai',
        'beijing', 'beijing', 'beijing'],

    'temperature' : [
        65, 66, 68,
        75, 78, 82,
        80, 77, 79],

    'humidity' : [
        56, 58, 60,
        80, 83, 85,
        26, 30, 35] })

df
```

```python
df.pivot(index='date', columns='city')
```

```python
df.pivot(index='date', columns='city', values='humidity')
```

# Pivot table

Pivot table is used to summarize and aggregate data inside a dataframe

```python
df1 = pd.DataFrame({
    'date' : [
        '5/1/2017', '5/1/2017', '5/2/2017', '5/2/2017',
        '5/1/2017', '5/1/2017', '5/2/2017', '5/2/2017'],
    'city' : [
        'new york', 'new york', 'new york', 'new york',
        'mumbai', 'mumbai', 'mumbai', 'mumbai'],
    'temperature' : [
        65, 61, 70, 72,
        75, 78, 82, 80],
    'humidity' : [
        56, 54, 60, 62,
        80, 83, 85, 26] })

df1
```

```python
df1.pivot_table(index='city', columns='date', aggfunc='mean')  # mean is default
```

```python
df1.pivot_table(index='city', columns='date', aggfunc='max') 
```

```python
df1.pivot_table(index='city', columns='date', aggfunc='mean', margins=True)

# Margins adds the All columns and rows which contains the mean of the 
# values it correlates to.
```

# Grouper in pivot table



```python
df2 = pd.DataFrame({
    'date' : [
        '5/1/2017', '5/3/2017', '5/3/2017',
        '12/1/2017', '12/3/2017', '12/3/2017'],
    'city' : [
        'new york', 'new york', 'new york', 
        'new york', 'new york', 'new york'],
    'temperature' : [
        65, 61, 70,
        30, 28, 25],
    'humidity' : [
        56, 54, 60,
        50, 52, 51] })
df2
```

```python
df2['date'] = pd.to_datetime(df2['date'])
df2
```

For Grouper freq Aliases see here:
    
https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#offset-aliases

```python
df2.pivot_table(index=pd.Grouper(freq='M', key='date'), columns='city')
```
