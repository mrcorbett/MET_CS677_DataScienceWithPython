```python
import pandas as pd

df = pd.read_csv('AAPL.csv', parse_dates=['Date'], index_col='Date')
df.head()
```

```python
df.index
# Note:  The index if of type DatetimeIndex which provides many benefits.
```

# Retrieve Jan 2017 stock prices

```python
df['2017-01']
```

# Find average price of stock in the month of Jan, 2017

```python
df['2017-01'].mean()
```

# Resampling

Convert daily data into monthly data

```python
# https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#dateoffset-objects

df['Close'].resample('M').mean()

# Shows the monthly averages of the closing price
```

```python
%matplotlib inline
#df['Close'].resample('M').mean().plot() # Monthly
df['Close'].resample('Q').mean().plot() # Quarterly
#df['Close'].resample('Q').mean().plot(kind='bar') # Quarterly
```
