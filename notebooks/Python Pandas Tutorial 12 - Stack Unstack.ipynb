### Read Excel spreadsheet with two layer column headers

```python
import pandas as pd

df = pd.read_excel('stocks-pandas tutorial 12 - stack unstack.xlsx', header=[0,1], index_col=[0])
df

# Header level 0: Price, Price to earnings ratio(P/E)
# Header level 1: Facebook, Google, Microsoft, Facebook, Google, Microsoft
# index_col=[0] gets rid of the Unnamed: 0_level_0
```

### Stack(level=0) Move level=0 column header to inner most index

```python
df1 = df.stack(level=0)
df1
```

```python
print(df1.index)  # default dropna=True
```

### Stack(level=1) Move level=1 column header to inner most index

```python
df.stack(level=1)  # default dropna=True
```

### Read Excel spreadsheed with three layer column headers

```python
df2 = pd.read_excel('stocks-pandas tutorial 12 - stack unstack 3 column headers.xlsx', header=[0,1, 2], index_col=[0])
df2
```

### Move level 2 column header to inner most index

```python
df3 = df2.stack(level=2)
df3
```

### Unstack the level 1 index back to the inner most column (=2)

```python
df3.unstack(level=1)
```
